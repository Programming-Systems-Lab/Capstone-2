```python
def select_available_features(X):
    # Delete useless features told by Shirish
    useless_features=['Package_Name', 'SHA256', 'ThumbPrint', 'Application_Category', 
            'Family_Name', 'Malware_Category', 'Malware/Benign','sdkVersion', 'targetSdkVersion','label']
    ## delete features with covariance=1
    drop_features = X.columns[X.nunique()==1]
    feature_lst_orig=set(X.columns)
    feature_lst=list(set(feature_lst_orig)-set(useless_features)-set(drop_features))
    # sensor related columns
    sensor_cols=X.loc[:, 'accelerometer':'temperature'].columns.to_list()
    sensor_cols = list(set(sensor_cols)-set(drop_features))
    print('--- Number of Sensor features: ',len(sensor_cols))
    print("--- Number of features (with sensors):", len(feature_lst))
    return feature_lst
```

## xgb

```python
def resample(df):
    gp=df.groupby('label').count().iloc[:,0:1].reset_index().sort_values('label')
    gp.columns=['label','cnt']
    cnt_lst=list(gp['cnt'])
    if cnt_lst[1]-cnt_lst[0]>0:
        x_df_resample=pd.concat([df.query('label==0').sample(n=cnt_lst[1]-cnt_lst[0],random_state=123,replace=True),df])
    else:
        x_df_resample=pd.concat([df.query('label==1').sample(n=cnt_lst[0]-cnt_lst[1],random_state=123,replace=True),df])
    y_df_resample=x_df_resample['label']
    return x_df_resample,y_df_resample
```

```python
def model_create_fit(optimizer,X_train,y_train,X_test,y_test):
    param_dct_tp=optimizer.max['params']
    param_dct={}
    for param,value in param_dct_tp.items():
        if param in {'max_depth','n_estimators'}:
            param_dct[param]=int(value)
        else:
            param_dct[param]=value
    print(param_dct)
    continuous = make_column_selector(dtype_exclude = 'object')
    model = xgb.XGBClassifier(**param_dct)
    model.fit(X_train, np.squeeze(y_train.to_numpy().reshape(1, -1)))
    y_pred = model.predict_proba(X_test)
    return model,y_pred
```

```python
results = pd.DataFrame()
# column names of final results
columns = ['upper_threshold', 'lower_threshold', 'train_size', 'test_size', 'percentage_malware', 'balanced accuracy',
          'benign_precision', 'malware_precision', 'benign_recall', 'malware_recall']

for lower_threshold in lower_thresholds:
    for upper_threshold in upper_thresholds:
        # only keep apps that are benign or malicious
        X_t = X[(X['Malware/Benign']<=lower_threshold)|(X['Malware/Benign']>=upper_threshold)]
        # get feature_lst
        feature_lst=select_available_features(X_t)
        
        # only keep apps that are benign or malicious
        X_train_t = X_train[(X_train['Malware/Benign']<=lower_threshold)|(X_train['Malware/Benign']>=upper_threshold)]
        # change labels accordingly
        y_train_t = np.array([1 if p>=upper_threshold else 0 for p in X_train_t['Malware/Benign']])
        # only keep apps that are benign or malicious
        X_test_t = X_test[(X_test['Malware/Benign']<=lower_threshold)|(X_test['Malware/Benign']>=upper_threshold)]
        # change labels accordingly
        y_test_t = np.array([1 if p>=upper_threshold else 0 for p in X_test_t['Malware/Benign']])
    
        X_train_resample,y_train_resample=resample(X_train_t)
    
        def xgbc_cv_with_sensor(max_depth,learning_rate,n_estimators,reg_alpha):
            estimator_function = xgb.XGBClassifier(max_depth=int(max_depth),
                                                   learning_rate= learning_rate,
                                                   n_estimators= int(n_estimators),
                                                   reg_alpha = reg_alpha,
                                                   nthread = -1,
                                                   objective='binary:logistic',
                                                   seed = seed)
            estimator_function.fit(X_train_resample[feature_lst],y_train_resample)
            probs = estimator_function.predict_proba(X_test_t[feature_lst])
            probs = probs[:,1]
            val1_roc = roc_auc_score(y_test_t,probs)
            return val1_roc
        
        
        warnings.filterwarnings('ignore')
        gp_params = {"alpha": 1e-10}
        seed = 112 # Random seed
        hyperparameter_space = {
            'max_depth': (5, 25),
            'learning_rate': (0, 1),
            'n_estimators' : (10,100),
            'reg_alpha': (0,1)
        }

        xgbcBO = BayesianOptimization(f = xgbc_cv_with_sensor, 
                                     pbounds =  hyperparameter_space,
                                     random_state = seed,
                                     verbose = 10)

        xgbcBO.maximize(init_points=3,n_iter=20,acq='ucb', kappa= 3, **gp_params)
        
        model_with_sensor,y_pred_with_sensor = model_create_fit(xgbcBO,X_train_resample[feature_lst],y_train_resample,X_test_t[feature_lst],y_test_t)
        del X_train_resample,y_train_resample
        
        # get predictions
        y_pred = model_with_sensor.predict(X_test_t[feature_lst])
        # calculate balanced accuracy
        bal_acc=balanced_accuracy_score(y_test_t, y_pred)
        # calculate precision
        benign_prec, malware_prec = precision_score(y_test_t, y_pred,average=None)
        # calculate recall
        benign_rec, malware_rec =  recall_score(y_test_t, y_pred,average=None)
        # obtain training data sample size
        train_size = len(y_train_t)
        # obtain test data sample size
        test_size = len(y_test_t)
        malware_percentage = (sum(y_train_t)+sum(y_test_t))/(len(y_train_t)+len(y_test_t))
    
        # add metrics to results
        df_threshold = pd.DataFrame([[upper_threshold, lower_threshold, train_size, test_size, malware_percentage, 
                                      bal_acc, benign_prec, malware_prec, benign_rec, malware_rec]],columns=columns)
        results = pd.concat([results, df_threshold])
        print("Upper:", upper_threshold, "Lower:", lower_threshold)
# reset index
results.reset_index(drop=True, inplace=True)
```

```python
import tensorflow as tf
from tensorflow import keras
import IPython
import kerastuner as kt
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import layers
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import ModelCheckpoint
from keras.losses import categorical_crossentropy as logloss
from keras.utils import np_utils
from keras.layers import Dropout
from keras.optimizers import adam
```

```python
## keras-tunner--- cannot add dropout layer ---- not useful enough
def model_builder(hp):
    tf.keras.backend.clear_session()
    model = keras.Sequential()
    # Tune the number of units in the first Dense layer
    # Choose an optimal value between 16-64
    hp_units1 = hp.Int('units1', min_value = 16, max_value = 64, step = 16)
    model.add(keras.layers.Dense(units = hp_units1, activation = 'relu'))
    model.add(keras.layers.Dropout(0.4))
    hp_units2 = hp.Int('units2', min_value = 16, max_value = 64, step = 16)
    model.add(keras.layers.Dense(units = hp_units2, activation = 'relu'))
    model.add(keras.layers.Dropout(0.4))
    model.add(keras.layers.Dense(1,activation='sigmoid'))

    # Tune the learning rate for the optimizer 
    # Choose an optimal value from 0.01, 0.001, or 0.0001
    hp_learning_rate = hp.Choice('learning_rate', values = [1e-1,1e-2, 1e-3, 1e-4]) 
    
#     model.compile(loss='binary_crossentropy', optimizer=adam(lr=1e-4, decay=1e-6), metrics=[tf.keras.metrics.AUC(),'accuracy'])

    model.compile(optimizer = keras.optimizers.Adam(learning_rate = hp_learning_rate),
                loss = keras.losses.BinaryCrossentropy(from_logits = True), 
                metrics=[tf.keras.metrics.Precision()])
    return model
```

```python
class ClearTrainingOutput(tf.keras.callbacks.Callback):
    def on_train_end(*args, **kwargs):
        IPython.display.clear_output(wait = True)
```

```python
results = pd.DataFrame()
# column names of final results
columns = ['upper_threshold', 'lower_threshold', 'train_size', 'test_size', 'percentage_malware', 'balanced accuracy',
          'benign_precision', 'malware_precision', 'benign_recall', 'malware_recall']

for lower_threshold in lower_thresholds:
    for upper_threshold in upper_thresholds:
        # only keep apps that are benign or malicious
        X_t = X[(X['Malware/Benign']<=lower_threshold)|(X['Malware/Benign']>=upper_threshold)]
        # get feature_lst
        feature_lst=select_available_features(X_t)
        
        # only keep apps that are benign or malicious
        X_train_t = X_train[(X_train['Malware/Benign']<=lower_threshold)|(X_train['Malware/Benign']>=upper_threshold)]
        # change labels accordingly
        y_train_t = np.array([1 if p>=upper_threshold else 0 for p in X_train_t['Malware/Benign']])
        X_train_t['label']=y_train_t
        # only keep apps that are benign or malicious
        X_test_t = X_test[(X_test['Malware/Benign']<=lower_threshold)|(X_test['Malware/Benign']>=upper_threshold)]
        # change labels accordingly
        y_test_t = np.array([1 if p>=upper_threshold else 0 for p in X_test_t['Malware/Benign']])
        X_test_t['label']=y_test_t
        X_train_resample,y_train_resample=resample(X_train_t)
    
        tf.keras.backend.clear_session()
#         tuner = kt.Hyperband(model_builder,
#                      kt.Objective("val_precision", direction="max"),
#                      max_epochs = 50,
#                      factor = 3,
#                      directory = 'dnn_model_training',
#                      project_name = '2_layers_wo_sensor',
#                      overwrite = True)   
        tuner = kt.tuners.bayesian.BayesianOptimization(model_builder, 
                                                 kt.Objective("val_precision", direction="max"),
                                                 max_trials=5, 
                                                 num_initial_points=2, 
                                                 seed=12, 
                                                 directory = 'dnn_model_training',
                                                 project_name = '2_layers_wo_sensor',
                                                 overwrite = True)

        tuner.search(X_train_resample[feature_lst].values,y_train_resample.values, epochs = 40,
             validation_data=(X_test_t[feature_lst].values,y_test_t),verbose=0,
             callbacks = [ClearTrainingOutput()])
        best_hps = tuner.get_best_hyperparameters(num_trials = 1)[0]
        tf.keras.backend.clear_session()
        model = keras.Sequential()
        model = Sequential()
        model.add(Dense(best_hps.get('units1'), input_dim=X_train_resample[feature_lst].shape[1], activation='relu'))
        model.add(Dropout(0.4))
        model.add(Dense(best_hps.get('units2'), activation='relu'))
        model.add(Dropout(0.4))
        # model.add(Dense(128, activation='relu'))
        model.add(Dense(1, activation='sigmoid'))

        # Compile model
        model.compile(loss='binary_crossentropy', optimizer=adam(lr=best_hps.get('learning_rate')), metrics=['accuracy',tf.keras.metrics.AUC()])
#         model.summary()
        history=model.fit(X_train_resample[feature_lst].values, y_train_resample,
                  validation_data=(X_test_t[feature_lst].values,y_test_t),
                  verbose=0,
                  epochs=50,batch_size=32)
        
        del X_train_resample,y_train_resample
        
        # get predictions
        y_pred = model.predict(X_test_t[feature_lst])
        y_pred=np.round(y_pred)
        # calculate balanced accuracy
        bal_acc=balanced_accuracy_score(y_test_t, y_pred)
        # calculate precision
        benign_prec, malware_prec = precision_score(y_test_t, y_pred,average=None)
        # calculate recall
        benign_rec, malware_rec =  recall_score(y_test_t, y_pred,average=None)
        # obtain training data sample size
        train_size = len(y_train_t)
        # obtain test data sample size
        test_size = len(y_test_t)
        malware_percentage = (sum(y_train_t)+sum(y_test_t))/(len(y_train_t)+len(y_test_t))
    
        # add metrics to results
        df_threshold = pd.DataFrame([[upper_threshold, lower_threshold, train_size, test_size, malware_percentage, 
                                      bal_acc, benign_prec, malware_prec, benign_rec, malware_rec]],columns=columns)
        results = pd.concat([results, df_threshold])
        print("Upper:", upper_threshold, "Lower:", lower_threshold)
# reset index
results.reset_index(drop=True, inplace=True)
```
